```python
# Create a relationship table where each hotel ID is matched with individual amenities
hotel_amenities = []

for index, row in df.iterrows():
    for amenity in row['amenities']:
        hotel_amenities.append({'ID': row['_id'], 'Amenity': amenity})

hotel_amenities_df = pd.DataFrame(hotel_amenities)

# Save the relationship table to a CSV
hotel_amenities_df.to_csv('hotel_amenities_mapping.csv', index=False)

print(len(hotel_amenities_df))
```
